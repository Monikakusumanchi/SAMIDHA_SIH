```python
import numpy as np
import pandas as pd
import seaborn as sns

import pandas as pd

import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("indexed-dataFrame.csv")
```

```python
df
```

```python
df.dropna(axis=0,inplace=True)
```

```python
df.loc[df['fuelConsumption'] < 0, 'fuelConsumption'] = 0
```

```python
df.describe()
```

```python
df['meanDraft'] = df[['draftAft', 'draftForward', 'draftMid1', 'draftMid2']].mean(axis=1)
```

```python
df1=df.drop(['draftAft', 'draftForward', 'draftMid1', 'draftMid2'], axis=1)
```

```python
cols_to_drop = ['HFO', 'MGO', 'currentDirection', 'currentSpeed', 'waterDepth', 'waveHeight', 'wavePeriod', 
                'waveDirection']
df1.drop(cols_to_drop, axis=1,inplace=True)
```

```python
df1.head(142)
```

```python
corr = df1.corr()
corr.style.background_gradient(cmap='coolwarm')
```

```python
X = df1.iloc[:,2:]
Y = df1.iloc[:,1]
```

```python
X
```

```python
Y
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaler.fit(X)
```

```python

standardized_data = scaler.transform(X)
```

```python
print(standardized_data)
```

```python
X = standardized_data
Y = df1['fuelConsumption']
```

```python
X
```

```python
Y
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2, random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
y_pred=model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import r2_score
Accuracy=r2_score(Y_test,y_pred)*100
print(" Accuracy of the model is %.2f" %Accuracy)
```
